Load & Glimpse

```python
import pandas as pd
```

```python
df=pd.read_csv("train.csv")
df.head()
```

```python
df.info()
```

Variable Categorization

```python
numerical_features = df.select_dtypes(include=["int64", "float64"]).columns.tolist()
categorical_features = df.select_dtypes(include=["object"]).columns.tolist()

print("Numerical Features:", numerical_features)
print("\nCategorical Features:", categorical_features)
```

Target Variable: SalePrice (Univariate Analysis)

```python
saleprice = df["SalePrice"]

mean_price = saleprice.mean()
median_price = saleprice.median()
skewness = saleprice.skew()
kurtosis = saleprice.kurt()

mean_price, median_price, skewness, kurtosis
```

Histogram,Q-Q Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
```

```python
plt.figure(figsize=(8,5))
sns.histplot(saleprice, kde=True)
plt.title("Histogram of SalePrice")
plt.show()
```

Q-Q Plot

```python
plt.figure(figsize=(6,6))
stats.probplot(saleprice, dist="norm", plot=plt)
plt.title("Q-Q Plot of SalePrice")
plt.show()
```

Log Transformation

```python
import numpy as np

df["SalePrice_log"] = np.log(df["SalePrice"])
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df["SalePrice_log"], kde=True)
plt.title("Histogram of Log(SalePrice)")
plt.show()
```

```python
plt.figure(figsize=(6,6))
stats.probplot(df["SalePrice_log"], dist="norm", plot=plt)
plt.title("Q-Q Plot of Log(SalePrice)")
plt.show()
```

IQR Calculation

```python
Q1 = saleprice.quantile(0.25)
Q3 = saleprice.quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

Q1, Q3, IQR, lower_bound, upper_bound
```

Outlier Filter

```python
df_cleaned = df[
    (df["SalePrice"] >= lower_bound) &
    (df["SalePrice"] <= upper_bound)
]

print("Before Cleaning:", df.shape)
print("After Cleaning:", df_cleaned.shape)
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(y=df["SalePrice"])
plt.title("Boxplot Before Outlier Removal")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(y=df_cleaned["SalePrice"])
plt.title("Boxplot After Outlier Removal")
plt.show()
```

Correlation Matrix & Heatmap

```python
import matplotlib.pyplot as plt
import seaborn as sns
corr_matrix = df[numerical_features].corr()
plt.figure(figsize=(14,10))
sns.heatmap(corr_matrix, cmap="coolwarm", annot=False)
plt.title("Correlation Heatmap of Numerical Features")
plt.show()
```

```python
top_corr = corr_matrix["SalePrice"].sort_values(ascending=False)
top_10_features = top_corr[1:11]
top_10_features
```

Categorical Deep-Dive

```python
plt.figure(figsize=(10,6))
sns.boxplot(x="OverallQual", y="SalePrice", data=df)
plt.title("OverallQual vs SalePrice")
plt.xlabel("Overall Quality (1â€“10)")
plt.ylabel("SalePrice")
plt.show()
```

Neighborhood Analysis

```python
neighborhood_price = df.groupby("Neighborhood")["SalePrice"].mean().sort_values(ascending=False)
plt.figure(figsize=(12,6))
neighborhood_price.plot(kind="bar")
plt.title("Average House Price by Neighborhood")
plt.ylabel("Average SalePrice")
plt.xlabel("Neighborhood")
plt.xticks(rotation=45)
plt.show()
```
